### ‚ùó **Factorial of a Number** üî¢

Imagine you have been hired as a Python developer for a math education üè´ startup. You have been assigned a task to develop a program that calculates the factorial of a positive integer using recursion.


Additionally, find the time and space complexity of the code you have developed.

factorial.png

````python
# Write your code here
def factorial(n):
  if n < 0:
    return "invalid number"
  elif n == 0:
    return 1
  else:
    return n * factorial(n-1)
````

````python
i = int(input("Enter the number here to get the factorial: "))
print(f"The factorial of {i} is {factorial(i)}")
````

### **Largest Integer in the List** üî¢

Consider a scenario where you are given a list of integers and you need to find the largest integer in the list using recursion in Python. 

Write a Python function called find_largest(lst) that takes a list of integers as input and returns the largest integer in the list using recursion.

For example, given the list `[3, 7, 1, 9, 4, 5]`, the find_largest function should return 9.


Additionally, find the time and space complexity of the code you have developed.

largest-element-of-array-in-python.png

````python
# Write your code here
def find_largest(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        return max(lst[0], find_largest(lst[1:]))
````

````python
num = [3, 7, 1, 9, 4, 5]
print(f"The largest number from the {num} is {find_largest(num)}")
````
